## Step 1: Get geo Data

```python
# I tried using geocoder but it would not work so I am using the CSV file provided
import pandas as pd
# use pandas read function to convert file to dataframe
geo_data_df = pd.read_csv("https://cocl.us/Geospatial_data")
geo_data_df.head()
```

```python
# same as above with the csv file created in part 1 of this exercise

toronto_df = pd.read_csv("toronto_base.csv")
toronto_df.head()
```

```python
# Merge two dataframes along the Postal Code using pandas merge function 

df_toronto_geo = pd.merge(toronto_df, geo_data_df, on="Postal Code")
df_toronto_geo.to_csv("toronto_geo.csv")
df_toronto_geo.head(80)
```
